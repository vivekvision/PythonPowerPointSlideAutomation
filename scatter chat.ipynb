```python
import pandas as pd
from pptx import Presentation
from pptx.chart.data import XyChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.util import Inches


"""
This code creates a scatter chart with lines connecting the points, 
using the XL_CHART_TYPE.XY_SCATTER_LINES chart type. The chart data is defined using XyChartData() 
and individual data points are added using add_data_point().
The resulting chart is added to a slide and the chart and axis titles are set before saving the presentation.

"""


# create sample dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [6, 8, 9, 4, 5]})

# create pptx presentation and slide
prs = Presentation()
slide = prs.slides.add_slide(prs.slide_layouts[5])

# define chart data
chart_data = XyChartData()
series = chart_data.add_series('Series 1')

for j in range(0, len(df)):
    series.add_data_point(df['x'][j], df['y'][j])


# add chart to slide
x, y, cx, cy = Inches(2), Inches(2), Inches(6), Inches(4.5)

chart = slide.shapes.add_chart(XL_CHART_TYPE.XY_SCATTER, x, y, cx, cy, chart_data).chart

# set chart title and axis titles
chart.has_title = True
chart.chart_title.text_frame.text = "XY Scatter Chart"
chart.category_axis.has_title = True
chart.category_axis.axis_title.text_frame.text = "X Axis"
chart.value_axis.has_title = True
chart.value_axis.axis_title.text_frame.text = "Y Axis"

# save presentation
prs.save('scatter_chart.pptx')
```

```python
import pandas as pd
from pptx import Presentation
from pptx.chart.data import XyChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.util import Inches


"""
This code creates a scatter chart with lines connecting the points, 
using the XL_CHART_TYPE.XY_SCATTER_LINES chart type. The chart data is defined using XyChartData() 
and individual data points are added using add_data_point().
The resulting chart is added to a slide and the chart and axis titles are set before saving the presentation.

"""



# Sample data
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y1': [10, 15, 13, 17, 12],
    'y2': [8, 13, 9, 14, 11]
})

# create pptx presentation and slide
prs = Presentation()
slide = prs.slides.add_slide(prs.slide_layouts[5])

# define chart data
chart_data = XyChartData()

series1 = chart_data.add_series('Series 1')
for j in range(0, len(df)):
    series1.add_data_point(df['x'][j], df['y1'][j])

series2 = chart_data.add_series('Series 2')
for j in range(0, len(df)):
    series1.add_data_point(df['x'][j], df['y2'][j])
    

    
# add chart to slide
x, y, cx, cy = Inches(2), Inches(2), Inches(6), Inches(4.5)

chart = slide.shapes.add_chart(XL_CHART_TYPE.XY_SCATTER, x, y, cx, cy, chart_data).chart

# set chart title and axis titles
chart.has_title = True
chart.chart_title.text_frame.text = "XY Scatter Chart"
chart.category_axis.has_title = True
chart.category_axis.axis_title.text_frame.text = "X Axis"
chart.value_axis.has_title = True
chart.value_axis.axis_title.text_frame.text = "Y Axis"

# save presentation
prs.save('multi_series_scatter_chart.pptx')
```
