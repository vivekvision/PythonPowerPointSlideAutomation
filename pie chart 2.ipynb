```python
from pptx import Presentation
from pptx.chart.data import CategoryChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.util import Inches

import pandas as pd

# create a new presentation
prs = Presentation()

# create a slide
slide = prs.slides.add_slide(prs.slide_layouts[5])

# create data for chart
data = pd.DataFrame({
    'category': ['Category 1', 'Category 2', 'Category 3'],
    'value': [10, 20, 30]
})

# create chart data
chart_data = CategoryChartData()
chart_data.categories = data['category']
chart_data.add_series('Series 1', data['value'])

# add chart to slide
left = Inches(2)
top = Inches(2)
width = Inches(6)
height = Inches(4.5)
chart = slide.shapes.add_chart(
    chart_type=XL_CHART_TYPE.PIE,
    x=left,
    y=top,
    cx=width,
    cy=height,
    chart_data=chart_data
).chart

# set chart title
chart.has_title = True
chart.chart_title.text_frame.text = 'Pie Chart'

# set data label position
chart.plots[0].has_data_labels = True
#chart.plots[0].data_labels.position = 'outside_end'

# save presentation
prs.save('pie_chart2.pptx')
```

```python
from pptx import Presentation
from pptx.chart.data import CategoryChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.util import Inches

import pandas as pd

# create a new presentation
prs = Presentation()

# create a slide
slide = prs.slides.add_slide(prs.slide_layouts[5])

# create data for chart
data = pd.DataFrame({
    'category': ['Category 1', 'Category 2', 'Category 3'],
    'value': [10, 20, 30]
})

# create chart data
chart_data = CategoryChartData()
chart_data.categories = data['category']
chart_data.add_series('Series 1', data['value'])

# add chart to slide
left = Inches(2)
top = Inches(2)
width = Inches(6)
height = Inches(4.5)
chart = slide.shapes.add_chart(
    chart_type=XL_CHART_TYPE.PIE,
    x=left,
    y=top,
    cx=width,
    cy=height,
    chart_data=chart_data
).chart

# set chart title
chart.has_title = True
chart.chart_title.text_frame.text = 'Pie Chart'

# add data labels to the chart series
series = chart.series[0]
data_labels = series.data_labels
data_labels.show_category = True
data_labels.show_value = True

# save presentation
prs.save('pie_chart3.pptx')
```
